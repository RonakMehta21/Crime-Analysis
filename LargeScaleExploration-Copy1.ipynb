```python
import pandas as pd
import numpy as np
from sklearn import preprocessing
from sklearn import tree
from sklearn.model_selection import train_test_split
from io import StringIO
from sklearn.tree import export_graphviz
from sklearn.metrics import classification_report,recall_score
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
```

```python
data=pd.read_csv('data.csv')
len(data)
```

```python
codes=pd.read_csv('codes.csv')
len(codes)

len(set(codes.CODE))
```

```python
result = pd.merge(data, codes, left_on='OFFENSE_CODE', right_on='CODE',how='left')
len(result)
```

```python
result2=result.drop_duplicates(subset=None, keep='first', inplace=False)
```

```python
result2=result2.drop(['INCIDENT_NUMBER','CODE','OFFENSE_CODE','OFFENSE_DESCRIPTION','OCCURRED_ON_DATE','Location','STREET'],axis=1)
```

```python
len(result2)
```

```python
result2.isnull().sum(axis=0)
```

```python
set(result2['SHOOTING'])
```

```python
result3=result2['SHOOTING'].fillna('0')
```

```python
result4=result3.replace('0',0)
```

```python
result4=result4.replace('1',1)
```

```python
result4=result4.replace('Y',1)
```

```python
set(result4)
```

```python
result2['SHOOTING']=result4
```

```python
result2.isnull().sum(axis=0)
```

```python
result_na=result2.dropna()
```

```python
result_na.isnull().sum(axis=0)
```

```python
result_na
```

```python
set(result2['REPORTING_AREA'])
```

```python
x2=x.drop('OFFENSE_CODE_GROUP', axis=1)
x2=x.drop('NAME', axis=1)
print(x2)
y=x.OFFENSE_CODE_GROUP
```

```python
model=tree.DecisionTreeClassifier(criterion="entropy")
xTrain, xTest, yTrain, yTest = train_test_split(x2, y, test_size=0.3, random_state=0)
```

```python
model.fit(xTrain,yTrain)
```

```python
model2 = model.predict(xTest)
#print(classification_report(yTest, model2))
print('The accuracy :- ',accuracy_score(yTest, model2)*100)
```

```python
neigh = KNeighborsClassifier(n_neighbors=3)
neigh.fit(x2, y)
```

```python
neigh2 = neigh.predict(xTest)
#print(classification_report(yTest, neigh2))
print('The accuracy :- ',accuracy_score(yTest, neigh2)*100)
```

```python
model=tree.DecisionTreeClassifier(criterion="gini")
xTrain, xTest, yTrain, yTest = train_test_split(x2, y, test_size=0.4, random_state=0)
model.fit(xTrain,yTrain)

model2 = model.predict(xTest)
#print(classification_report(yTest, model2))
print('The accuracy :- ',accuracy_score(yTest, model2)*100)
```
